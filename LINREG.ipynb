```python
import numpy as np
import matplotlib.pyplot as plt

# Linear regression function
def linreg(X, y):
    X_bias = np.concatenate((X, -np.ones((X.shape[0], 1))), axis=1)  # Add bias term
    return np.linalg.pinv(X_bias) @ y

# Define inputs and targets
inputs = np.array([[0,0], [0,1], [1,0], [1,1]])

testin = np.concatenate((inputs, -np.ones((inputs.shape[0], 1))), axis=1)  # Add bias term
ANDtargets = np.array([[0], [0], [0], [1]])
ORtargets = np.array([[0], [1], [1], [1]])
XORtargets = np.array([[0], [1], [1], [0]])

# Perform linear regression for each gate
print("AND data")
ANDbeta = linreg(inputs, ANDtargets)
ANDout = np.dot(testin, ANDbeta)
print(ANDout)

print("OR data")
ORbeta = linreg(inputs, ORtargets)
ORout = np.dot(testin, ORbeta)
print(ORout)

print("XOR data")
XORbeta = linreg(inputs, XORtargets)
XORout = np.dot(testin, XORbeta)
print(XORout)

# Plotting results
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Plot for AND gate
axs[0].scatter(inputs[:, 0], inputs[:, 1], c=ANDtargets.flatten(), cmap='bwr', s=100, label="AND")
axs[0].set_title("AND Gate")
axs[0].set_xlabel("Input 1")
axs[0].set_ylabel("Input 2")
axs[0].set_xlim([-0.5, 1.5])
axs[0].set_ylim([-0.5, 1.5])
axs[0].legend()

# Plot for OR gate
axs[1].scatter(inputs[:, 0], inputs[:, 1], c=ORtargets.flatten(), cmap='bwr', s=100, label="OR")
axs[1].set_title("OR Gate")
axs[1].set_xlabel("Input 1")
axs[1].set_ylabel("Input 2")
axs[1].set_xlim([-0.5, 1.5])
axs[1].set_ylim([-0.5, 1.5])
axs[1].legend()

# Plot for XOR gate
axs[2].scatter(inputs[:, 0], inputs[:, 1], c=XORtargets.flatten(), cmap='bwr', s=100, label="XOR")
axs[2].set_title("XOR Gate")
axs[2].set_xlabel("Input 1")
axs[2].set_ylabel("Input 2")
axs[2].set_xlim([-0.5, 1.5])
axs[2].set_ylim([-0.5, 1.5])
axs[2].legend()

plt.tight_layout()
plt.show()

```
